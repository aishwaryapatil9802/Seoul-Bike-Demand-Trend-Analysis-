```python
### . Encode Categorical Variables ###
# Convert Yes/No columns to 1/0
df['Functioning Day'] = df['Functioning Day'].map({'Yes': 1, 'No': 0})
df['Holiday'] = df['Holiday'].map({'Yes': 1, 'No': 0})
```

## **Data Visualizations with charts:**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Hourly Bike Rental Demand
plt.figure(figsize=(8, 5))
hourly_demand = df.groupby("Hour")["Rented Bike Count"].mean()
sns.lineplot(x=hourly_demand.index, y=hourly_demand.values, marker="o", color="b")
plt.xlabel("Hour of the Day")
plt.ylabel("Average Rented Bike Count")
plt.title("Hourly Bike Rental Demand")
plt.xticks(range(0, 24))
plt.grid(True)
plt.show()
```

**Graph:** Line Plot of Bike Rentals Over Time.

   **Insight:** Identify seasonal trends, peak rental months, and overall demand fluctuations.
The highest demand for bike rentals occurs between 6 AM - 9 AM (morning commute) and 5 PM - 8 PM (evening commute).
Lowest demand occurs between midnight and early morning hours.

   **Business Use:** Helps in forecasting demand and adjusting bike availability.

```python
# 2.Rental bikes across the seasons

plt.figure(figsize=(8, 5))
sns.barplot(x=df['Seasons'], y=df['Rented Bike Count'],hue=df["Seasons"],palette="bright")
plt.xlabel("Season")
plt.ylabel("Bike Rentals")
plt.title("Bike Rentals Across Seasons")
plt.show()

```

**Graph:** Boxplot or Bar Chart of Rentals per Season.

**Insight:** Understand how demand changes across spring, summer, autumn, and winter.
Highest demand in Spring & Autumn (pleasant weather).
Lowest demand in Winter due to cold and snow.

**Business Use:** Plan promotions in low-demand seasons & maintenance schedules.

```python
# 3. Impact of Temperature on Bike Rentals
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["Temperature(°C)"], y=df["Rented Bike Count"], alpha=0.5, color="b")
plt.xlabel("Temperature (°C)")
plt.ylabel("Rented Bike Count")
plt.title("Impact of Temperature on Bike Rentals")
plt.show()
```

 **Graph:** Scatter Plot (Temperature vs. Rentals) + Bar Charts for Rainfall/Snowfall.
 
**Insight:** Shows how temperature, humidity, wind speed, and rainfall affect demand.Bike rentals increase with temperature, peaking around 15-25°C.
Extreme temperatures (very hot or very cold) reduce bike usage.

 **Business Use:** Plan dynamic pricing and special offers on extreme weather days.




```python
# 4.Rainfall vs. Bike Rentals

plt.figure(figsize=(10, 5))
sns.scatterplot(x=df['Rainfall(mm)'], y=df['Rented Bike Count'], color='black')
plt.xlabel("Rainfall (mm)")
plt.ylabel("Bike Rentals")
plt.title("Impact of Rainfall on Bike Rentals")
plt.show()
```

```python
# 5.Humidity vs. Bike Rentals ###
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Humidity(%)'], y=df['Rented Bike Count'], alpha=0.6, color='green')
plt.xlabel("Humidity (%)")
plt.ylabel("Bike Rentals")
plt.title("Impact of Humidity on Bike Rentals")
plt.show()
```

Heavy rain and high humidity significantly decrease bike rentals.

Most users avoid biking on rainy days.

```python
# 6. Bike rentals on Holidays Vs. Non-Holidays

plt.figure(figsize=(6, 5))
sns.barplot(x=df['Holiday'], y=df['Rented Bike Count'],hue=df['Holiday'],palette="coolwarm")
plt.xlabel("Holiday (Yes/No)")
plt.ylabel("Bike Rentals")
plt.title("Bike Rentals on Holidays vs. Non-Holidays")
plt.show()
```

**Graph:** Boxplot or Bar Chart (Bike Rentals on Holidays vs. Regular Days)

**Insight:** Shows whether people rent more on holidays.Fewer rentals on public holidays compared to working days. Bike usage drops on non-functioning days (weekends & holidays)

**Business Use:** Offer discounts on holidays or increase fleet availability.

```python
# 7. Impact of Visibility on Bike Rentals
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Visibility (10m)"], y=df["Rented Bike Count"], alpha=0.5, color="purple")
plt.xlabel("Visibility (10m)")
plt.ylabel("Rented Bike Count")
plt.title("Impact of Visibility on Bike Rentals")
plt.show()
```

```python
# 8. Impact of Wind Speed on Bike Rentals

plt.figure(figsize=(8, 5))
sns.scatterplot(x=df["Wind speed (m/s)"], y=df["Rented Bike Count"], alpha=0.8, color="red")
plt.xlabel("Wind Speed (m/s)")
plt.ylabel("Rented Bike Count")
plt.title("Impact of Wind Speed on Bike Rentals")
plt.show()

```

**Graph:** Scatter Plots (Visibility vs. Rentals, Wind Speed vs. Rentals)

**Insight:** Understand if low visibility or high wind speed reduces bike rentals. Poor visibility (fog) reduces rentals as safety concerns rise.

Strong winds also decrease bike usage.

**Business Use:** Helps in safety measures & predicting low-usage days.

```python
# 9. Bike Rentals on Functioning vs. Non-Functioning Days\
plt.figure(figsize=(6, 5))
functioning_day_demand = df.groupby("Functioning Day")["Rented Bike Count"].mean()
sns.barplot(x=functioning_day_demand.index, y=functioning_day_demand.values, color="green",legend=False)
plt.xlabel("Functioning Day Status")
plt.ylabel("Average Rented Bike Count")
plt.title("Bike Rentals on Functioning vs. Non-Functioning Days")
plt.show()
```

**Graph:** Bar Chart of Rentals on Open vs. Closed Days

**Insight:** See how demand is affected on non-functioning days (e.g., maintenance or public holidays).Rentals are significantly higher on functioning days (working days) compared to non-functioning days (holidays or maintenance days). This confirms that bike rentals are primarily used for commuting rather than leisure.

**Business Use:** Plan maintenance schedules efficiently.

```python
# 10. Heat map

plt.figure(figsize=(10, 6))
numeric_df = df.select_dtypes(include=['number'])
# Generate the heatmap
sns.heatmap(numeric_df.corr(), annot=True, cmap='magma', fmt='.2f', linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```

**Graph:** Heatmap of Feature Correlation

**Insight:** Identify which factors (e.g., temperature, humidity, hour, season) impact rentals the most. 

Positive correlation: Temperature has a strong positive impact on bike rentals (warmer temperatures increase rentals).

Negative correlation: Rainfall, humidity, and wind speed negatively impact rentals (bad weather reduces demand).

Weak correlation: Visibility has a smaller effect on rentals compared to other weather factors.

**Business Use:** Helps in building accurate prediction models for demand forecasting.
